```python
import pandas as pd
import numpy as np

prices = pd.read_csv(
    "../data/raw/sp500_stocks.csv",
    index_col=0,
    parse_dates=True
)

market = pd.read_csv(
    "../data/raw/sp500_index.csv",
    index_col=0,
    parse_dates=True
)

rf = pd.read_csv(
    "../data/raw/risk_free.csv",
    index_col=0,
    parse_dates=True
)
```

```python
print(prices.shape)
print(prices.head())

print(market.head())

print(rf.head())

print(prices.isna().sum())
```

```python
prices.isna().sum().sort_values(ascending=False)
```

```python
stock_returns = prices.pct_change().dropna()

market_returns = market.pct_change().dropna()

market_returns.columns = ["Market"]
```

```python
rf = rf / 100
```

```python
daily_rf = (1 + rf) ** (1/252) - 1
```

```python
daily_rf.columns = ["RF"]
```
