```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasets/s-and-p-500-companies/master/data/constituents.csv"

sp500_table = pd.read_csv(url)

print(sp500_table.head())
```

```python
TICKERS = {
    "Technology": ["AAPL", "MSFT", "NVDA", "GOOGL", "ORCL"],
    "Financials": ["JPM", "GS", "V", "MA", "BAC"],
    "Healthcare": ["LLY", "JNJ", "ABBV", "MRK", "PFE"],
    "Energy": ["XOM", "CVX", "COP", "SLB", "EOG"],
    "Consumer": ["WMT", "COST", "MCD", "KO", "PEP"],
    "Industrials": ["GE", "RTX", "CAT", "HON", "UNP"]
}
```

```python
tickers = []

for sector in TICKERS.values():
    tickers.extend(sector)

print(tickers)
print(len(tickers))
```

```python
import yfinance as yf

START = "2015-01-01"
END = "2025-12-31"

prices = yf.download(
    tickers,
    start=START,
    end=END,
    auto_adjust=True,
    progress=True
)["Close"]
```

```python
market = yf.download(
    "^GSPC",
    start=START,
    end=END,
    auto_adjust=True,
    progress=False
)["Close"]
```

```python
rf = yf.download(
    "^IRX",
    start=START,
    end=END,
    auto_adjust=False,
    progress=False
)["Close"]
```

```python
prices.to_csv("../data/raw/sp500_stocks.csv")
market.to_csv("../data/raw/sp500_index.csv")
rf.to_csv("../data/raw/risk_free.csv")
```
